```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset('titanic')
```

```python
data.info()
```

```python
data.head(15)
```

```python
data.isna().sum()
```

```python
sns.boxplot(data)
```

```python
sns.countplot(data=data,x='survived')
plt.xticks([0, 1], ['Not-Survived', 'Survived'])
plt.show()
```

```python
sns.countplot(data=data,x='survived',hue='sex')
plt.xticks([0, 1], ['Not-Survived', 'Survived'])
plt.show()
```

```python
sns.countplot(data=data,x='survived',hue='class')
plt.xticks([0, 1], ['Not-Survived', 'Survived'])
plt.show()
```

```python
data['class'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
sns.boxplot(data['fare'])
```

```python
sns.barplot(data=data,y='age',x='class')#by default estimator is average
```

```python
sns.barplot(data=data,y='age',x='class',estimator='sum')
```

```python
sns.barplot(data=data,y='age',x='class',estimator='sum',hue='sex')
```
