```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('googleplaystore.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['Rating']=data['Rating'].fillna(data['Rating'].mean())
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.head()
```

### Q1.How many free apps are there in Art & Design

```python
df=data.copy()
```

```python
df[df['Genres']=='Art & Design']
```

```python
df_=df[df['Category']=='ART_AND_DESIGN' ]
len(df_[df_['Type']=='Free'])##there are 54 apps with free type
```

```python
len(df[(df['Category']=='ART_AND_DESIGN')&(df['Type']=='Free')])
```

### Q2.How many apps are there in Art & Design with rating more than 4.5?

```python
len(df[(df['Category']=='ART_AND_DESIGN')&(df['Rating']>=4.5)])
```

### Q3.How many apps are there in FAMILY with more than 4.5 rating and Free?

```python
len(df[(df['Category']=='FAMILY') & (df['Type']=='Free')&(df['Rating']>=4.5)])
```

### Q4.List all apps with category Family and with more than 4.5 rating and Free type

```python
df[(df['Category']=='FAMILY') & (df['Type']=='Free')&(df['Rating']>=4.5)]
```

### Q5.Sort all the apps with Category ART_AND_DESIGN and more than 4.5 rating in ascending order WRT rating 

```python
df[(df['Category']=='ART_AND_DESIGN')&(df['Rating']>=4.5)].sort_values(by='Rating',ascending=True)
```

### Q6.Which category have maximum average rating?

```python
df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
```
